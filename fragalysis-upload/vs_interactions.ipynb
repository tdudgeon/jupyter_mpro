```python
# read fields from file

supplr = Chem.ForwardSDMolSupplier(FILE)

def add_inter(prop, name):
    if prop.startswith(name + ':'):
        pos = prop.find(':')
        inter_type = prop[:pos]
        canon_res = prop[pos + 1:]
        desc = 'Score for {} {} interaction'.format(canon_res, inter_type)
        print('Adding', desc)
        mol0.SetProp(prop, desc)
        
missing = set()
count = 0
for mol in supplr:
    count += 1
    if not mol:
        print("WARNING: failed to read molecule {}".format(count))
        continue
    for prop in mol.GetPropNames():
        if not mol0.HasProp(prop):
            add_inter(prop, 'HydrogenBond')
            add_inter(prop, 'HalogenBond')
            add_inter(prop, 'Hydrophobic')
            add_inter(prop, 'SaltBridge')
            add_inter(prop, 'PiStacking')
            add_inter(prop, 'PiCation')
            if not mol0.HasProp(prop) and prop not in missing:
                print('WARNING: property', prop, 'not described')
                missing.add(prop)

print('Read {} molecules'.format(count))
```

```python
supplr = Chem.ForwardSDMolSupplier(FILE)

w = Chem.SDWriter(OUTPUT)
w.write(mol0)
count = 0
for mol in supplr:
    count += 1
    if not mol:
        print("WARNING: failed to read molecule {}".format(count))
    w.write(mol)
    
w.close()
print('Wrote', count, 'molecules')
```
